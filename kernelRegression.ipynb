```python
import pandas as pd
import matplotlib.pyplot as plt

filepath = "data/tempexam.csv"
temp_data = pd.read_csv(filepath, sep=";")

temp_data.head()
temp_data.columns

```

```python
plt.figure(figsize=(10,6))
plt.plot(temp_data['Dates'], temp_data['y'], marker='o', color='red', kind="line")
plt.xlabel('Dates')
plt.ylabel('Temperature')
plt.title('Temperature')
plt.show()
```
